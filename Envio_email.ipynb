```python
import os
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
from datetime import datetime

def encontrar_arquivo_excel(diretorio, nome_arquivo):
    """Busca um arquivo Excel específico em um diretório."""
    for root, dirs, files in os.walk(diretorio):
        for file in files:
            if file.endswith('.xlsm') and nome_arquivo in file:
                return os.path.join(root, file)
    return None

def enviar_email_com_anexo(email_de, senha, email_para, assunto, mensagem, caminho_arquivo):
    """Envia um e-mail com um anexo."""
    # Configuração do e-mail
    msg = MIMEMultipart()
    msg['From'] = email_de
    msg['To'] = email_para
    msg['Subject'] = assunto

    # Corpo do e-mail
    msg.attach(MIMEText(mensagem, 'plain'))

    # Anexo
    try:
        with open(caminho_arquivo, 'rb') as arquivo:
            parte_anexo = MIMEBase('application', 'octet-stream')
            parte_anexo.set_payload(arquivo.read())
            encoders.encode_base64(parte_anexo)
            parte_anexo.add_header('Content-Disposition', f'attachment; filename={os.path.basename(caminho_arquivo)}')
            msg.attach(parte_anexo)
    except Exception as e:
        print(f"Erro ao anexar o arquivo: {e}")
        return

    # Conexão com o servidor SMTP e envio
    try:
        servidor = smtplib.SMTP('smtp.gmail.com', 587)
        servidor.starttls()
        servidor.login(email_de, senha_de)
        texto = msg.as_string()
        servidor.sendmail(email_de, email_para, texto)
        servidor.quit()
        print(f"E-mail enviado com sucesso para {email_para}")
    except smtplib.SMTPAuthenticationError:
        print("Erro de autenticação: verifique seu e-mail e senha.")
    except smtplib.SMTPException as e:
        print(f"Erro no envio do e-mail: {e}")


email = "jonatas.tavares@metropoles.com"
senha = "djvu rsvu efai ncym"
data_atual = datetime.now().strftime('%d/%m/%Y')

if __name__ == "__main__":
    diretorio = "C://Users//Jonatas.Tavares//Documents//Planilha"
    nome_arquivo = "Vendas_final"  # Parte do nome do arquivo para procurar

    email_de = email
    senha_de = senha
    email_para = "gestaocomercial@metropoles.com"
    assunto = f"BACKUP - Planilha de vendas - {data_atual}"
    mensagem = f"""Boa noite, Rafa, Frâncio e Priscila.

Segue anexa planilha Excel de vendas a título de BACKUP o dia. A mesma será salva também no drive. 

Abraço,"""

    # Buscar o arquivo
    caminho_arquivo = encontrar_arquivo_excel(diretorio, nome_arquivo)

    if caminho_arquivo:
        # Enviar o e-mail
        enviar_email_com_anexo(email_de, senha_de, email_para, assunto, mensagem, caminho_arquivo)
        print("Mensagem enviada!")
    else:
        print(f"Arquivo {nome_arquivo} não encontrado em {diretorio}")
```
